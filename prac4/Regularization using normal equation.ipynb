```python
import numpy as np
import pandas as pd
from numpy.linalg import inv
from sklearn.datasets import load_boston
```

```python
X,y = load_boston(return_X_y=True)
```

```python
X_train = X[0:400, : ]

temp = np.zeros((X_train.shape[0], X_train.shape[1] + 1))
temp[:,0] = np.ones(X_train.shape[0])
temp[:,1:] = X_train
X_train = temp
print(X_train.shape)
```

```python
y_train = y[0:400]
y_train.shape
```

```python
lamba = 10
mat = np.identity(X_train.shape[1])
mat[0,0] = 0
print(mat)
```

```python
theta = np.zeros(X_train.shape[1])
print(theta)
```

```python
XTXi = inv(np.dot(X_train.T,X_train) + lamba*mat)
XTXi
```

```python
XTy = np.dot(X_train.T,y_train)
XTy
```

```python
theta = np.dot(XTXi,XTy)
theta
```

```python
prediction = np.dot(theta,X_train.T)
prediction
```
