# PDF OCR

```python
from pdf2image import convert_from_path 
import os
```

Snapshot of the input pdf file

![title](img/img1.png)

```python
pdf_input = convert_from_path('robert_frost.pdf', 500) 

counter = 1
for page in pdf_input: 
    filename = str(counter)+".jpg"                #Each page is turned to an image to process 
    page.save(filename, 'JPEG')                   #image is saved in the format of jpeg
    
    page_img = Image.open(filename)

    text = pytesseract.image_to_string(page_img)
                     
    print("For page",counter)
    print(text)
    counter = counter + 1
```

# Limitations -

1. Images need to be scaled appropriately

2. have as much image contrast as possible

3. the text must be horizontally aligned.

4. Tesseract isn’t trained to recognize handwriting, and it’s limited to about 100 fonts in total

5. doesnt give output when the image is very small text

6. not very accurate in giving the output for rotated or skewed input

7. background should be plain with no lining or shapes

8. works best when text is 300DPI(300 dots per inch) thus if this isnt the case with the input text we have to resize it
