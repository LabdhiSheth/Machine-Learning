Tabula is a simple Python wrapper of tabula-java. It can read tables from PDF and convert into pandas's DataFrame. It also provides the functionality to convert a PDF file into CSV/TSV/JSON file.

```python
import tabula
import pandas as pd
```

__read_pdf__
some of the useful features of this method are:

1. input_path - path of the source file

2. output_format - the format of the returned object like dataframe

3. multiple_tables - this is uused to handle multiple tables in a page in the pdf, default value is true.it used pandas datab=frame for the tables

4. pages - specify the page number to extract the tabble from or it can be "all"

5. stream - if stream is true then force PDF to be extracted using stream-mode extraction



```python
table = tabula.read_pdf('data.pdf')
print(len(table))
```

```python
table[0]
```

```python
tabula.read_pdf('data.pdf', pages = "all")
```

__convert_into__
this function is used to convert tables from pdf to a specific file format.

some of its important features are:

1. input_path - path of the pdf to extract the tables from

2. output_path - name of the output file

3. output_format - extension of the output file, example - csv or json

4. pages - page from which we wish to extract, an be "all"

5. stream - if stream is true then force PDF to be extracted using stream-mode extraction

6. area - area under the table in pdf units, useful when extracting a single page from a page of multile tables

```python
tabula.convert_into('data.pdf', 'table1.csv',pages = '1', output_format='csv', stream = True)
data_table = pd.read_csv('table1.csv')
data_table.head()
```

```python
tabula.convert_into('data.pdf', 'table2.csv',pages = '2', output_format='csv', stream = True)
```

__export__

exports the list of tables to a specified format

Parameters - 

1. path(str) - output file path

2. f(str) - file format can be csv,json,excel,html

3. compress(boolean) - whether or not to add to zip archive



```python
table.export('camelot_tables_data.csv', f='csv', compress=True)
```

## Comparison

Condition|Tabula|Camelot
-|-|-
When the header text is vertical|doesn't output all header|outputs all headers
column spans multiple cells|tabula outputs all the headers but moves some headers to incorrect positions|camelot gives correct output
row spans multiple cells|shifts the data giving wrong output|gives correct output with data as it is in correct cells
a page with multiple tables|we have to explicity output each table to different csv/json/excel file|all the tables are exported in different csv/json/excel files implicitly
zip file archive| can not compress to zip while converting|export function compressess to zip folder
table with figures and text|doesn't give an output|gives output by preserving table structure
rotated table(all the texts are vertical)|gives completely wrong output|gives correct output with text written horizontly



Inference - Camelot library has a better performance ratio in comparison to tabula in most of the cases.
